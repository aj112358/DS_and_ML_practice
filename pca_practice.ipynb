```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer["DESCR"])
```

```python
df = pd.DataFrame(data=cancer["data"], columns=cancer["feature_names"])
```

```python
df
```

We will now start our PCA. The first step is normalizing the data w.r.t. variance

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_df = pd.DataFrame(data=scaled_data, columns=cancer["feature_names"])
scaled_df
```

Now we do the actual PCA analysis/algorithm

```python
pca_object = PCA(n_components=2)
```

```python
pca_object.fit(scaled_data)
```

```python
x_pca = pca_object.transform(scaled_data)
```

```python
scaled_data.shape, x_pca.shape
# So...we have reduced our data set to have simply the TWO most important features!
```

```python
x_pca
```

```python
# let's now visualize our data set w.r.t only these two PCs

plt.figure(figsize=(10,6))
plt.scatter(x_pca[:,0], x_pca[:,1], c=cancer["target"])
plt.xlabel("First PC")
plt.ylabel("Second PC")
```

We can also investigate and visualize the individual PC, and how much they are affected by the original features.

```python
pca_object.components_
```

```python
comp_df = pd.DataFrame(data=pca_object.components_, columns=cancer["feature_names"])
```

```python
comp_df
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(comp_df, cmap="plasma")
```

Since we only care about the absolute value of the measurements in the heatmap...

```python
plt.figure(figsize=(14,6))
sns.heatmap(np.abs(comp_df))
```
