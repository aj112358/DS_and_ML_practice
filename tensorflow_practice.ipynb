```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import load_model
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df = pd.read_csv("TensorFlow_FILES/DATA/fake_reg.csv")
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
X = df[["feature1", "feature2"]].values
type(X)
```

```python
X
```

```python
y = df["price"].values
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

Creating Model

```python
model = Sequential()
```

```python
model.add(Dense(units=4, activation="relu")) # Input Layer
model.add(Dense(units=4, activation="relu"))
model.add(Dense(units=4, activation="relu"))

model.add(Dense(units=1)) # Output Layer
```

```python
model.compile(optimizer="rmsprop", loss="mse")
```

```python
model.fit(x=X_train, y=y_train, epochs=250)
```

```python
model.history.history
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df
```

```python
loss_df.plot(figsize=(10,5))
```

Evaluate the Model

```python
model.evaluate(X_test, y_test)
```

```python
# model.evaluate(X_train, y_train)
```

```python
predictions = model.predict(X_test)
predictions
```

```python
predictions = pd.DataFrame(predictions, columns=["Predictions"])
predictions
```

```python
y_test = pd.DataFrame(y_test, columns=["True_Values"])
y_test
```

```python
predict_df = pd.concat([y_test, predictions], axis=1)
```

```python
predict_df
```

```python
sns.scatterplot(x="True_Values", y="Predictions", data=predict_df)
```

```python
mean_absolute_error(predict_df["True_Values"], predict_df["Predictions"])
```

```python
df.describe()
```

So, we have a MAE of \\$4.00. According to our 'price' data above, the average price is \\$498.67. So, we see that our MAE is under 1% of this average price, which is pretty good! This good/low error is also indicated qualitatively through our scatter plot above.

Thus, we can conclude that our model is performing very well at predicting the price based on the two original features!

```python
mean_squared_error(predict_df["True_Values"], predict_df["Predictions"])
# As expected, it's the same as what we got from the NN history above
```

```python
new_features_scaled = scaler.transform(new_features)
new_features_scaled
```
